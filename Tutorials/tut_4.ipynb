```python
import matplotlib.pyplot as plt
import numpy as np

dt = 0.01
init_t = 0
max_t = 5

tarr = np.linspace(init_t, max_t, round((max_t-init_t)/dt))
```

```python
def get_arr(init_N, K, r) :

    N = init_N
    N_arr = []

    for elem in tarr :
        N_arr.append(N)
        N += dt * (r * N * (1 - N/K))

    return N_arr
```

```python
plt.figure(figsize=(8, 5))
plt.plot(tarr, get_arr(1, 10, 1))
plt.plot(tarr, get_arr(10, 10, 1))
plt.plot(tarr, get_arr(20, 10, 1))
plt.plot(tarr, get_arr(1, 10, 2))
plt.plot(tarr, get_arr(5, 10, 2))
plt.plot(tarr, get_arr(20, 10, 2))
# plt.plot(tarr, get_arr(100, 10, 3))
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
 
X, Y = np.meshgrid(x, y)
 
u = np.ones((100, 100)) 
v = get_arr(Y, X, 1)
print(v)
 
fig = plt.figure(figsize = (8, 5))
plt.streamplot(X, Y, u, v, density = 0.5)
plt.show()
```
