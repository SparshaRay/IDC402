# The logistic map
##### (With the Feigenbaum Tree)

~ inspired by https://virtualmathmuseum.org/Fractal/feigenbaum_tree/index.html

```python
import numpy as np

num_x, num_r = 37, 1000

x = np.linspace(0.01, 0.99, num_x)
r = np.linspace(0, 4, num_r)

xval, rval = np.meshgrid(x, r)
```

```python
import matplotlib.cm as cm

c = cm.hsv(x)
c[:, 3] = 0.5

cols = []

for i in range(num_r) : 
    for elem in c :
        cols.append(elem)
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')

iters = 5

for i in range(iters) :
    xval = rval * xval * (1 - xval)

plt.figure(figsize=(16, 8))
plt.scatter(rval, xval, color=cols, s=1)
plt.show()
```
