# Renormalization

```python
import matplotlib.pyplot as plt
import numpy as np

def func(x, r):
    return r - x**2

r = 0
alpha = -2
x = np.linspace(-1, 1, 100)
diag = np.linspace(-1, 1, 100)

plt.figure(figsize=(8, 5))
plt.plot(x, diag)
plt.plot(x, func(x, r))
r = 1
plt.plot(x, func(func(x, r), r))
plt.plot(x, alpha*func(func(x/alpha, r), r))
plt.show()
```
