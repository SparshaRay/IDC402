# Lyapunov exponent

```python
def f(x, r):
    return r * x * (1-x)

def fder(x, r):
    return r - 2 * r * x
```

```python
import numpy as np

r = 0.01
stepsize = 0.001
iters = 500
threshold = 100

xarr = []
rarr = []
lyapunov_coeff = []
lyapunov_r = []


for i in range(round(4/stepsize)):
    x = 0.5
    lpnvsum = 0
    for i in range(iters):
        x = f(x, r)
        if i>threshold :
            xarr.append(x)
            rarr.append(r)
            lpnvsum += np.log(np.abs(fder(x, r)))
    lyapunov_coeff.append(lpnvsum/iters)
    lyapunov_r.append(r)
    r += stepsize

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(25,10))
# plt.scatter(rarr, xarr, s=0.1, color = 'black', alpha = 0.1)
plt.scatter(lyapunov_r, lyapunov_coeff, s=0.1, color = 'red')
plt.ylim(-2, 1)
plt.show()
```
