# The tent map

```python
def tent_map(x) :
    if x < 0.5 :
        return 1.99*x
    else :
        return 1.99*(1-x)
    

r = 3.85
def logistic_map(x):
    return r * x * (1 - x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 5))
plt.ylim = (0, 1)
xrange = np.linspace(0, 1, 1000)
yval = list(map(logistic_map, xrange))
plt.plot(xrange, yval)
plt.show()
```

#### and it goes on like this if $x_n+2$ is plotted against $x_n$

```python
plt.figure(figsize=(8, 5))
plt.ylim = (0, 1)

for iter in range(3) :
    yval = list(map(logistic_map, yval))

plt.plot(xrange, yval)
plt.show()
```

# Now the timeseries plots

```python
initx = 0.3
num_points = 20
r = 2.95

currentx = initx
xarr = []
yarr = list(np.arange(0, num_points))

for num in yarr :
    xarr.append(currentx)
    currentx = logistic_map(currentx)

plt.figure(figsize=(10, 5))
plt.ylim = [(0, 1)]
plt.plot(yarr, xarr, 'r-')

print(f'Converges to {xarr[-1]} for r = {r}')

initx = 0.3
r = 2.9

currentx = initx
xarr = []
yarr = list(np.arange(0, num_points))

for num in yarr :
    xarr.append(currentx)
    currentx = logistic_map(currentx)

plt.plot(yarr, xarr, 'b-')
plt.show()
```
